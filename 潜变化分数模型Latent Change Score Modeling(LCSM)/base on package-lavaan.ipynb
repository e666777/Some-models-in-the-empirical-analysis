```r
DATA = read.csv("DATA.csv")
library(lavaan)
```

```r
model1.uni_x <-
      "
      # Specify latent true scores 
      lx1 =~ 1 * x1
      lx2 =~ 1 * x2
      lx3 =~ 1 * x3
      lx4 =~ 1 * x4
      lx5 =~ 1 * x5
      lx6 =~ 1 * x6
      # Specify mean of latent true scores 
      lx1 ~ gamma_lx1 * 1 
      lx2 ~ 0 * 1 
      lx3 ~ 0 * 1 
      lx4 ~ 0 * 1 
      lx5 ~ 0 * 1 
      lx6 ~ 0 * 1
      # Specify variance of latent true scores 
      lx1 ~~ sigma2_lx1 * lx1 
      lx2 ~~ 0 * lx2 
      lx3 ~~ 0 * lx3 
      lx4 ~~ 0 * lx4 
      lx5 ~~ 0 * lx5 
      lx6 ~~ 0 * lx6 
      # Specify intercept of obseved scores 
      x1 ~ 0 * 1
      x2 ~ 0 * 1
      x3 ~ 0 * 1
      x4 ~ 0 * 1
      x5 ~ 0 * 1
      x6 ~ 0 * 1
      # Specify variance of observed scores 
      x1 ~~ sigma2_ux * x1
      x2 ~~ sigma2_ux * x2
      x3 ~~ sigma2_ux * x3
      x4 ~~ sigma2_ux * x4
      x5 ~~ sigma2_ux * x5
      x6 ~~ sigma2_ux * x6
      # Specify autoregressions of latent variables 
      lx2 ~ 1 * lx1
      lx3 ~ 1 * lx2
      lx4 ~ 1 * lx3
      lx5 ~ 1 * lx4
      lx6 ~ 1 * lx5
      # Specify latent change scores 
      dx2 =~ 1 * lx2 
      dx3 =~ 1 * lx3 
      dx4 =~ 1 * lx4 
      dx5 =~ 1 * lx5 
      dx6 =~ 1 * lx6 
      # Specify latent change scores means 
      dx2 ~ 0 * 1
      dx3 ~ 0 * 1
      dx4 ~ 0 * 1
      dx5 ~ 0 * 1
      dx6 ~ 0 * 1
      # Specify latent change scores variances 
      dx2 ~~ 0 * dx2
      dx3 ~~ 0 * dx3
      dx4 ~~ 0 * dx4
      dx5 ~~ 0 * dx5
      dx6 ~~ 0 * dx6
      # Specify constant change factor 
      g2 =~ 1 * dx2 + 1 * dx3 + 1 * dx4 + 1 * dx5  + 1 * dx6
      # 恒定变化的均值 
      g2 ~ alpha_g2 * 1 
      # Specify constant change factor variance 
      g2 ~~ sigma2_g2 * g2 
      # Specify constant change factor covariance with the initial true score 
      g2 ~~ sigma_g2lx1 * lx1
      # Specify proportional change component 
      dx2 ~ beta_x * lx1 
      dx3 ~ beta_x * lx2 
      dx4 ~ beta_x * lx3 
      dx5 ~ beta_x * lx4 
      dx6 ~ beta_x * lx5 
      # Specify autoregression of change score 
      dx3 ~ phi_x * dx2 
      dx4 ~ phi_x * dx3 
      dx5 ~ phi_x * dx4 
      dx6 ~ phi_x * dx5
      "
#gamma_lx1	潜在真实分数的平均值 x （截距）
#sigma2_lx1	潜在真实分数的方差 x
#sigma2_ux	观测分数的方差 x
#alpha_g2	变化因子均值 （g2）
#alpha_g3	变化因子均值 （g3）
#sigma2_g2	变化因子方差 （g2）
#sigma2_g3	变化因子方差 （g3）
#sigma_g2lx1	变化因子 （g2） 的协方差与 初始真实分数 x
#sigma_g3lx1	变化因子 （g3） 的协方差与 初始真实分数 x
#sigma_g2g3	变化因子的协方差 构造 X
#beta_x	比例变化 x
#phi_x	变化分数的自回归 x

fit1 <- sem(model1.uni_x,DATA)
summary(fit1,fit.measures = TRUE,standardize = TRUE,rsquare =TRUE)
fitmeasures(fit1)
```

```r
model.bi <- 
    "
    # Specify parameters for construct x ----
    # Specify latent true scores 
    lx1 =~ 1 * x1 
    lx2 =~ 1 * x2 
    lx3 =~ 1 * x3 
    lx4 =~ 1 * x4 
    lx5 =~ 1 * x5 
    lx6 =~ 1 * x6
    # Specify mean of latent true scores 
    lx1 ~ 21 * 1 
    lx2 ~ 0 * 1 
    lx3 ~ 0 * 1 
    lx4 ~ 0 * 1 
    lx5 ~ 0 * 1
    lx6 ~ 0 * 1 
    # Specify variance of latent true scores 
    lx1 ~~ 0.5 * lx1 
    lx2 ~~ 0 * lx2 
    lx3 ~~ 0 * lx3 
    lx4 ~~ 0 * lx4 
    lx5 ~~ 0 * lx5 
    lx5 ~~ 0 * lx6
    # Specify intercept of obseved scores 
    x1 ~ 0 * 1 
    x2 ~ 0 * 1 
    x3 ~ 0 * 1 
    x4 ~ 0 * 1 
    x5 ~ 0 * 1 
    x6 ~ 0 * 1
    # Specify variance of observed scores 
    x1 ~~ 0.2 * x1
    x2 ~~ 0.2 * x2
    x3 ~~ 0.2 * x3
    x4 ~~ 0.2 * x4
    x5 ~~ 0.2 * x5
    x6 ~~ 0.2 * x6
    # Specify autoregressions of latent variables 
    lx2 ~ 1 * lx1
    lx3 ~ 1 * lx2
    lx4 ~ 1 * lx3
    lx5 ~ 1 * lx4
    lx6 ~ 1 * lx5
    # Specify latent change scores 
    dx2 =~ 1 * lx2
    dx3 =~ 1 * lx3
    dx4 =~ 1 * lx4
    dx5 =~ 1 * lx5
    dx6 =~ 1 * lx6
    # Specify latent change scores means 
    dx2 ~ 0 * 1 
    dx3 ~ 0 * 1 
    dx4 ~ 0 * 1 
    dx5 ~ 0 * 1 
    # Specify latent change scores variances 
    dx2 ~~ 0 * dx2 
    dx3 ~~ 0 * dx3 
    dx4 ~~ 0 * dx4 
    dx5 ~~ 0 * dx5 
    # Specify constant change factor 
    g2 =~ 1 * dx2 + 1 * dx3 + 1 * dx4 + 1 * dx5 + 1 * dx6 
    # Specify constant change factor mean 
    g2 ~ -0.4 * 1 
    # Specify constant change factor variance 
    g2 ~~ 0.4 * g2 
    # Specify constant change factor covariance with the initial true score 
    g2 ~~ 0.2 * lx1
    # Specify proportional change component 
    dx2 ~ -0.1 * lx1
    dx3 ~ -0.1 * lx2
    dx4 ~ -0.1 * lx3
    dx5 ~ -0.1 * lx4
    dx6 ~ -0.1 * lx5
    # Specify parameters for construct y ----
    # Specify latent true scores 
    ly1 =~ 1 * y1
    ly2 =~ 1 * y2
    ly3 =~ 1 * y3
    ly4 =~ 1 * y4
    ly5 =~ 1 * y5
    ly6 =~ 1 * y6
    # Specify mean of latent true scores 
    ly1 ~ 5 * 1
    ly2 ~ 0 * 1
    ly3 ~ 0 * 1
    ly4 ~ 0 * 1
    ly5 ~ 0 * 1
    ly6 ~ 0 * 1
    # Specify variance of latent true scores 
    ly1 ~~ 0.2 * ly1 
    ly2 ~~ 0 * ly2
    ly3 ~~ 0 * ly3
    ly4 ~~ 0 * ly4
    ly5 ~~ 0 * ly5
    ly6 ~~ 0 * ly6
    # Specify intercept of obseved scores 
    y1 ~ 0 * 1
    y2 ~ 0 * 1
    y3 ~ 0 * 1
    y4 ~ 0 * 1
    y5 ~ 0 * 1
    y6 ~ 0 * 1
    # Specify variance of observed scores 
    y1 ~~ 0.2 * y1
    y2 ~~ 0.2 * y2
    y3 ~~ 0.2 * y3
    y4 ~~ 0.2 * y4
    y5 ~~ 0.2 * y5
    y6 ~~ 0.2 * y6
    # Specify autoregressions of latent variables 
    ly2 ~ 1 * ly1
    ly3 ~ 1 * ly2
    ly4 ~ 1 * ly3
    ly5 ~ 1 * ly4
    ly6 ~ 1 * ly5
    # Specify latent change scores 
    dy2 =~ 1 * ly2
    dy3 =~ 1 * ly3
    dy4 =~ 1 * ly4
    dy5 =~ 1 * ly5
    dy6 =~ 1 * ly6
    # Specify latent change scores means 
    dy2 ~ 0 * 1
    dy3 ~ 0 * 1
    dy4 ~ 0 * 1
    dy5 ~ 0 * 1
    dy6 ~ 0 * 1
    # Specify latent change scores variances 
    dy2 ~~ 0 * dy2
    dy3 ~~ 0 * dy3
    dy4 ~~ 0 * dy4
    dy5 ~~ 0 * dy5
    dy6 ~~ 0 * dy6
    # Specify constant change factor 
    j2 =~ 1 * dy2 + 1 * dy3 + 1 * dy4 + 1 * dy5 
    # Specify constant change factor mean 
    j2 ~ -0.2 * 1 
    # Specify constant change factor variance 
    j2 ~~ 0.1 * j2 
    # Specify constant change factor covariance with the initial true score 
    j2 ~~ 0.02 * ly1
    # Specify proportional change component 
    dy2 ~ -0.2 * ly1
    dy3 ~ -0.2 * ly2
    dy4 ~ -0.2 * ly3
    dy5 ~ -0.2 * ly4
    dy6 ~ -0.2 * ly5
    # Specify autoregression of change score 
    dy3 ~ 0.1 * dy2
    dy4 ~ 0.1 * dy3
    dy5 ~ 0.1 * dy4
    dy6 ~ 0.1 * dy5
    # Specify residual covariances 
    x1 ~~ 0.01 * y1
    x2 ~~ 0.01 * y2
    x3 ~~ 0.01 * y3
    x4 ~~ 0.01 * y4
    x5 ~~ 0.01 * y5
    x6 ~~ 0.01 * y6
    # Specify covariances betweeen specified change components (alpha) and intercepts (initial latent true scores lx1 and ly1) ----
    # Specify covariance of intercepts 
    lx1 ~~ 0.2 * ly1 
    # Specify covariance of constant change and intercept within the same construct 
    ly1 ~~ 0.1 * g2 
    # Specify covariance of constant change and intercept within the same construct 
    lx1 ~~ 0.1 * j2 
    # Specify covariance of constant change factors between constructs 
    g2 ~~ 0.01 * j2 
    # Specify between-construct coupling parameters ----
    # Change score x (t) is determined by true score y (t-1)  
    dx2 ~ 0.13 * ly1
    dx3 ~ 0.13 * ly2
    dx4 ~ 0.13 * ly3
    dx5 ~ 0.13 * ly4
    dx6 ~ 0.13 * ly5
    # Change score y (t) is determined by change score x (t-1)  
    dy3 ~ 0.4 * dx2
    dy4 ~ 0.4 * dx3
    dy5 ~ 0.4 * dx4
    dy6 ~ 0.4 * dx5
    "
#gamma_lx1	潜在真实分数的平均值 x （截距）
#sigma2_lx1	潜在真实分数的方差 x
#sigma2_ux	观测分数的方差 x
#alpha_g2	变化因子均值 （g2）
#alpha_g3	变化因子均值 （g3）
#sigma2_g2	变化因子方差 （g2）
#sigma2_g3	变化因子方差 （g3）
#beta_x	比例变化 x
#sigma_g2lx1	变化因子 （g2） 的协方差与 初始真实分数 X （LX1）
#sigma_g3lx1	变化因子 （g3） 的协方差与 初始真实分数 X （LX1）
#sigma_g2g3	变化因子的协方差 构造 X
#phi_x	变化分数的自回归 x
#构造 Y	
#gamma_ly1	潜在真分数 y 的平均值 （截距）
#sigma2_ly1	潜在真实分数的方差 y
#sigma2_uy	观测分数的方差 y
#alpha_j2	变化因子均值 （j2）
#alpha_j3	变化因子均值 （j3）
#sigma2_j2	变化因子方差 （j2）
#sigma2_j3	变化因子方差 （j3）
#beta_y	比例变化 y
#sigma_j2ly1	变化因子 （j2） 的协方差与 初始真实分数 y （ly1）
#sigma_j3ly1	变化因子 （j3） 的协方差与 初始真实分数 y （ly1）
#sigma_j2j3	变化因子的协方差 构造 Y
#phi_y	变化分数的自回归 y
#Coupeling X & Y	
#sigma_su	残差 x 和 y 的协方差
#sigma_ly1lx1	截距 x 和 y 的协方差
#sigma_g2ly1	变化因子 x （g2） 的协方差与 初始真实分数 y （ly1）
#sigma_g3ly1	变化因子 x （g3） 的协方差与 初始真实分数 y （ly1）
#sigma_j2lx1	变化因子 y （j2） 的协方差与 初始真实分数 x （lx1）
#sigma_j3lx1	变化因子 y （j3） 的协方差与 初始真实分数 x （lx1）
#sigma_j2g2	变化因子 y （j2） 和 x 的协方差 （G2）
#sigma_j2g3	变化因子 y （j2） 和 x 的协方差 （G3）
#sigma_j3g2	变化因子 y （j3） 和 x 的协方差 （G2）
#delta_con_xy	变化分数 x （t） 由 true 确定 分数 y （t）
#delta_con_yx	变化分数 y （t） 由 true 确定 得分 x （t）
#delta_lag_xy	变化分数 x （t） 由 true 确定 分数 Y （T-1）
#delta_lag_yx	变化分数 y （t） 由 true 确定 分数 x （t-1）
#xi_con_xy	变化分数 x （t） 由变化决定 分数 y （t）
#xi_con_yx	变化分数 y （t） 由变化决定 得分 x （t）
#xi_lag_xy	变化分数 x （t） 由变化决定 分数 Y （T-1）
#xi_lag_yx	变化分数 y （t） 由变化决定 分数 x （t-1）
fit2 <- sem(model.bi,DATA)
summary(fit2,fit.measures = TRUE,standardize = TRUE,rsquare =TRUE)
fitmeasures(fit2)
```
